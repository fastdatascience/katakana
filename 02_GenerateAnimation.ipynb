```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("assets/katakana_a_white.png" )
```

```python
image.shape
```

```python
plt.imshow(image)
```

```python
import json
training_examples = []
with open("a.txt", "r", encoding="utf-8") as f:
    for l in f:
        training_examples.append(json.loads(l))
```

```python
objects = training_examples[0]['objects']
paths = [o for o in objects if o['type'] == 'path']
```

```python
import numpy as np
```

```python
frame_no = 0
radius = 20


for path_no, path in enumerate(paths): # iterate over strokes
    points = path['path']

    ctr = 0

    for i in range(0, len(points)):
        x_cen = points[i][1]
        y_cen = points[i][2]
        
        x=int(np.round(x_cen))
        y=int(np.round(y_cen))
        
        img2 = np.copy(image)

        cv2.circle(img2, (x,y), 30, (255,0,0), 10)
        plt.imshow(img2)
        cv2.imwrite(f"frame{frame_no:04d}.png", img2)

        frame_no += 1
```
